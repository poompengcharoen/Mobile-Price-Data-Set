```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
df = pd.read_csv('archive/train.csv')
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## 1. How does price affect ram?

```python
sns.jointplot(x = 'ram', y = 'price_range', data = df, kind = 'kde', height = 8)
```

```python
stats.pearsonr(df.ram, df.price_range)
```

Pearson correlation is 0.92 which mean that price and ram are positively correlated.

## 2. How are price range and internal memory related?

```python
sns.pointplot(x = 'price_range', y = 'int_memory', data = df)
```

The higher the price the more internal memory there is, with the exception of price range 2 which seems to be lower than price range 1 and 0.

## 3. What is the ratio of phones with 3G?

```python
labels = ['3G-supported', 'Not supported']
values = df['three_g'].value_counts().values
fig1, ax1 = plt.subplots()
ax1.pie(values, labels = labels, autopct = '%1.1f%%', shadow = True, startangle = 90)
```

Most phones support 3G, as 76.2% of them do.

## 4. What is the ratio of phones with 4G?

```python
labels4g = ['4G-supported','Not supported']
values4g = df['four_g'].value_counts().values
fig1, ax1 = plt.subplots()
ax1.pie(values4g, labels = labels4g, autopct = '%1.1f%%', shadow = True, startangle = 90)
```

About half of all the phones support 4G.

## 5. How are price range and battery power related?

```python
sns.boxplot(x = 'price_range', y = 'battery_power', data = df)
```

```python
stats.pearsonr(df.battery_power, df.price_range)
```

Pearson correlation is 0.2 which mean that price and battery power are positively correlated. The higher the phone price, the higher the battery power.

## Train model with logistic regression model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x = df.drop('price_range', axis = 1)
y = df['price_range']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 101)
```

```python
logmodel = LogisticRegression()
logmodel.fit(x_train, y_train)
print('The accuracy of the logistic regression model is', logmodel.score(x_test, y_test))
```
